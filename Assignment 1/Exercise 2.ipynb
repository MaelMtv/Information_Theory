```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
series = []
with open("temperatures.txt") as f:
    line = f.readline()
    while line != '':
        series.append(int(float(line)))
        line = f.readline()
```

```python
outcomes = set(series)
prob = {x:0.0 for x in outcomes}
d, h, counter = {}, [], 0

for i in range(len(series)):
    counter += 1
    d[series[i]]= d.get(series[i], 0) + 1#Increment the number of times the value appears
    for k, v in d.items():
        prob[k] = d.get(k)/counter#Compute new probabilities
    h.append( sum( - v*np.log2(v) for v in prob.values() if v != 0) )#Compute new entropy
    
print('Possible outcomes:',outcomes)
print('Number of values in each bin:',d)
print('Probabilities of each bin:',prob)
plt.plot(h)
plt.title('Evolution of entropy of the average temperature on earth each year')
plt.xlabel('Number of years since 1875 we consider')
plt.ylabel('Entropy')
plt.show()
```
